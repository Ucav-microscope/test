```python
Location = [] #빈 리스트 생성 
for row in a['Embarked'] : # 탑승장소의 값을 가져옴
    if row == 'S': 
        Location.append('S_Location') # 탑승장소에 맞는 값을들을 새로 넣음 
    elif row == 'C': 
        Location.append('C_Location') 
    else : 
        Location.append('Q_Location') 
            
            
            
a['Location'] = Location  # Location 열에 순차대로 원소를 대입

# 데이터가 크면 클수록 오래 걸릴 것 같다. 
```
